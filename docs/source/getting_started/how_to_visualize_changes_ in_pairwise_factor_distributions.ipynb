```python
import numpy as np
import pandas as pd
from skillmodels.config import TEST_DIR
import yaml
from skillmodels.simulate_data import simulate_dataset
from skillmodels.visualize_factor_distributions import plot_pairwise_factor_distributions
```

# How to visualize changes in pairwise factor distributions

We show the Kernel density plots for pairs of factors for simulated data with and without policy at certain period in gird or seperate plots. 

The function works both for list of simulated datasets and one simulated dataset. 


Below we show how to visualize pairwise changes in factor distributions for a test model. 

## Getting simulated data 
For more details on this check out the how to simulate dataset documentation. 

```python
with open(TEST_DIR/"model2.yaml") as y:
        model_dict = yaml.load(y, Loader=yaml.FullLoader)

params = pd.read_csv(TEST_DIR / "regression_vault" / f"one_stage_anchoring.csv")
params = params.set_index(["category", "period", "name1", "name2"])
```

```python
control_means = pd.Series([0], index=["x1"])
control_sds = pd.Series([1], index=["x1"])
```

##  Simulated data without policy

```python
initial_data =simulate_dataset(model_dict, params, n_obs=100, control_means=control_means,
                             control_sds=control_sds, policies=None)
```

```python
policies = [
        {"period": 1, "factor": "fac1", "effect_size": 3.5, "standard_deviation": 0.0},
        {"period": 1, "factor": "fac2", "effect_size": 3.5, "standard_deviation": 0.0}]
```

##  Simulated data with policy

```python
data_after_policies = simulate_dataset(model_dict, params, n_obs=100, control_means=control_means,
                             control_sds=control_sds, policies=policies)
```

```python
states=[initial_data ,data_after_policies]
```

##  Plot pairwise factor distributions in 2D as a grid for changes in initial_data and data_after_policies




```python
plot_pairwise_factor_distributions(
    states=states, 
    model_dict=model_dict,
    period=1,
    combined_plots=True,
    add_3d_plots=None,
    state_ranges=None,
    n_points=50)
```

##  Plot pairwise factor distributions in 2D as a sepeate plots for changes in initial_data and data_after_policies






```python
plot_pairwise_factor_distributions(
    states=states, 
    model_dict=model_dict,
    period=1,
    combined_plots=None,
    add_3d_plots=None,
    state_ranges=None,
    n_points=50)

```

## Plot pairwise factor distributions in 2D as a grid for one simulated dataset





```python
one_state=initial_data
```

```python
plot_pairwise_factor_distributions(
    states=one_state, 
    model_dict=model_dict,
    period=0,
    combined_plots=True,
    add_3d_plots=None,
    state_ranges=None,
    n_points=50)
```
